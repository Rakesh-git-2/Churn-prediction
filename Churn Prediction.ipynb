```python
X['gender'] = data['gender'].map({'Female':0,'Male':1})
X['Partner'] = data['Partner'].map({'Yes':1,'No':0})
X['Dependents'] = data['Dependents'].map({'Yes':1,'No':0})
X['PhoneService'] = data['PhoneService'].map({'Yes':1,'No':0})
X['MultipleLines'] = data['MultipleLines'].map({'No phone service':0, 'No':0, 'Yes':1})
X['InternetService'] = data['InternetService'].map({'DSL':1, 'Fiber optic':2, 'No':0})
X['OnlineSecurity'] = data['OnlineSecurity'].map({'Yes':1,'No':0,'No internet service':0})
X['OnlineBackup'] = data['OnlineBackup'].map({'Yes':1,'No':0,'No internet service':0})
X['DeviceProtection'] = data['DeviceProtection'].map({'Yes':1,'No':0,'No internet service':0})
X['TechSupport	'] = data['TechSupport'].map({'Yes':1,'No':0,'No internet service':0})
X['StreamingMovies'] = data['StreamingMovies'].map({'Yes':1,'No':0,'No internet service':0})
X['StreamingTV'] = data['StreamingTV'].map({'Yes':1,'No':0,'No internet service':0})
X['PaperlessBilling'] = data['PaperlessBilling'].map({'Yes':1,'No':0,'No internet service':0})
X['PaymentMethod'] = data['PaymentMethod'].map({'Electronic check':1, 'Mailed check':2, 'Bank transfer (automatic)':3,
       'Credit card (automatic)':4})
X['Contract'] = data['Contract'].map({'Month-to-month':1, 'One year':2, 'Two year':3})
X['MonthlyCharges'] = data['MonthlyCharges'].apply(lambda x : np.nan_to_num(x))
X['TotalCharges'] = data['TotalCharges'].apply(lambda x : np.nan_to_num(x))


 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy for Logistic Regression is {}%".format(accuracy_score(y_pred,y_test)*100) )
```
